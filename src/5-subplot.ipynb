### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np



x_line = np.arange(0, 11)
y_line = x_line ** 2


categories_bar = ['A', 'B', 'C', 'D']
values_bar = [5, 7, 3, 9]
colors_bar = ['red', 'green', 'blue', 'orange']


x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]


labels_pie = ['Python', 'Java', 'C++', 'JavaScript']
sizes_pie  = [40, 25, 20, 15]


fig, axes = plt.subplots(2, 2, figsize=(12, 9))


axes[0, 0].plot(x_line, y_line, marker='o')
axes[0, 0].set_title('Line Chart: $y = x^2$')
axes[0, 0].set_xlabel('x')
axes[0, 0].set_ylabel('y')
axes[0, 0].grid(True)


axes[0, 1].bar(categories_bar, values_bar, color=colors_bar)
axes[0, 1].set_title('Bar Chart')
axes[0, 1].set_xlabel('Category')
axes[0, 1].set_ylabel('Value')


axes[1, 0].scatter(x_scatter, y_scatter)
axes[1, 0].set_title('Scatter Plot')
axes[1, 0].set_xlabel('x')
axes[1, 0].set_ylabel('y')
axes[1, 0].grid(True)


axes[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', startangle=90)
axes[1, 1].set_title('Pie Chart')


plt.tight_layout()
plt.show()
```
